```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Consideramos una población que no puede comprender más de N = 10000000 individuos y definimos las tasas de natalidad y mortalidad (tambien migracion):
N = 10000000  # maximum population size
a = .049   # birth rate
b = .016   # death rate
m = .008   #migration rate
```

```python
# Simulamos una cadena de Markov en el espacio finito 0,1, ..., N. 
#Cada estado representa un tamaño de población. El vector x contendrá el tamaño de la población en cada paso de tiempo. 
#Establecemos el estado inicial en x0 = 6200000 (es decir, hay 6,200,000 individuos en la población en el momento de la inicialización):

nsteps = 540
x = np.zeros(nsteps)
x[0] = 6200000
```

```python
#Ahora simulamos nuestra cadena. En cada paso de tiempo t, hay un nuevo nacimiento con probabilidad axt, e independientemente, 
#hay una nueva muerte con probabilidad bxt. Estas probabilidades son proporcionales al tamaño de la población en ese momento. 
#Si el tamaño de la población llega a 0 o N, la evolución se detiene:

for t in range(nsteps - 1):
    if 0 < x[t] < N - 1:
        # Is there a birth?
        birth = np.random.rand() <= a * x[t]
        # Is there a death?
        death = np.random.rand() <= b * x[t]
        # Is there a migration?
        migr = np.random.rand() <= m * x[t]
        # We update the population size.
        x[t + 1] = x[t] + 303 * birth - 102 * death + 54 * migr
    # The evolution stops if we reach $0$ or $N$.
    else:
        x[t + 1] = x[t]
```

```python
#Observamos la evolución del tamaño de la población:
fig, ax = plt.subplots(1, 1, figsize=(8, 4))
ax.plot(x, lw=2)
```
